1. **ডেটা সংগ্রহ এবং একীকরণ:** 

- এমএল পাইপলাইনের প্রথম ধাপে ডেটা সংগ্রহ এবং ডেটা একীকরণ জড়িত।
- সংগৃহীত ডেটা মডেলের ইনপুট হিসাবে কাজ করে (ডেটা প্রস্তুতি পর্ব) 
- ইনপুটকে ফিচার বলা হয়।
- আমাদের বিবেচিত উদাহরণের ক্ষেত্রে সংগৃহীত ডেটাতে প্রচুর ডেটা জড়িত। সংগৃহীত তথ্য নিম্নলিখিত প্রশ্নের উত্তর দিতে হবে- অতীত গ্রাহক ইতিহাস কি? অতীত আদেশ কি ছিল? গ্রাহক কি আমাদের বইয়ের দোকানের প্রধান সদস্য? গ্রাহক কি কিন্ডলের মালিক? গ্রাহক কি কোনো পূর্বের অভিযোগ করেছেন? অভিযোগের সংখ্যা সবচেয়ে বেশি কি ছিল?
ডেটা যত বেশি হবে, আমাদের মডেল তত বেশি উন্নত হবে।
- একবার ডেটা সংগ্রহ করা হলে আমাদেরকে একত্রিত করতে হবে এবং ডেটা প্রস্তুত করতে হবে।
- ডেটার ইন্টিগ্রেশন মানে সমস্ত সম্পর্কিত ডেটা একসাথে রাখা।
- তারপরে ডেটা প্রস্তুতি পর্ব শুরু হয় যেখানে আমরা ম্যানুয়ালি এবং সমালোচনামূলকভাবে ডেটা অন্বেষণ করি।
- ডেটা প্রস্তুতি পর্বটি বিকাশকারীকে বলে যে ডেটা প্রত্যাশার সাথে মিলে যায়৷ একটি সঠিক ভবিষ্যদ্বাণী করতে যথেষ্ট তথ্য আছে? তথ্য সামঞ্জণ ডেটা এবং পরীক্ষার ডেটার উপর ভবিষ্যদ্বাণী করা হয়।

2. **অনুসন্ধানমূলক ডেটা বিশ্লেষণ এবং ভিজ্যুয়ালাইজেশন:**

- একবার ডেটা প্রস্তুত হয়ে গেলে ডেভেলপারকে ডেটাসেটের মধ্যে সম্পর্কের আরও ভাল বোঝার জন্য ডেটা ভিজ্যুয়ালাইজ করতে হবে।
- যখন আমরা ডেটা দেখতে পাই, তখন আমরা অদেখা নিদর্শনগুলি লক্ষ্য করতে পারি যা আমরা প্রথম পর্যায়ে লক্ষ্য করিনি।
- এটি বিকাশকারীদের সহজেই অনুপস্থিত ডেটা এবং বহিরাগতদের সনাক্ত করতে সহায়তা করে।
- হিস্টোগ্রাম, স্ক্যাটার প্লট ইত্যাদি প্লট করে ডেটা ভিজ্যুয়ালাইজেশন করা যেতে পারে।
- ভিজ্যুয়ালাইজেশন সম্পন্ন হওয়ার পরে ডেটা বিশ্লেষণ করা হয় যাতে বিকাশকারী সিদ্ধান্ত নিতে পারে যে সে কোন এমএল কৌশল ব্যবহার করতে পারে।
- বিবেচিত উদাহরণের ক্ষেত্রে গ্রাহকের ক্রয় অভ্যাস বিশ্লেষণ করার জন্য তত্ত্বাবধানহীন শিক্ষা ব্যবহার করা যেতে পারে।


3. **বৈশিষ্ট্য নির্বাচন এবং প্রকৌশল:**

- বৈশিষ্ট্য নির্বাচন মানে বিকাশকারী মডেলের মধ্যে কোন বৈশিষ্ট্যগুলি ব্যবহার করতে চান তা নির্বাচন করা।
- বৈশিষ্ট্যগুলি নির্বাচন করা উচিত যাতে তাদের মধ্যে একটি সর্বনিম্ন সম্পর্ক বিদ্যমান থাকে এবং নির্বাচিত বৈশিষ্ট্য এবং আউটপুটের মধ্যে একটি সর্বাধিক পারস্পরিক সম্পর্ক বিদ্যমান থাকে।
- ফিচার ইঞ্জিনিয়ারিং হল মূল ডেটাকে নতুন এবং সম্ভাব্য ডেটাতে ম্যানিপুলেট করার প্রক্রিয়া যার মধ্যে অনেকগুলি বৈশিষ্ট্য রয়েছে।
- সহজ কথায় ফিচার ইঞ্জিনিয়ারিং হল কাঁচা ডেটাকে দরকারী ডেটাতে রূপান্তর করা বা আসল ডেটা থেকে সর্বাধিক পাওয়া।
- ফিচার ইঞ্জিনিয়ারিং যুক্তিযুক্তভাবে এমএল পাইপলাইনের সবচেয়ে গুরুত্বপূর্ণ এবং সময়সাপেক্ষ পদক্ষেপ।
- বৈশিষ্ট্য নির্বাচন এবং প্রকৌশল প্রশ্নের উত্তর - এই বৈশিষ্ট্যগুলি কি আমাদের ভবিষ্যদ্বাণীতে কোন অর্থ বহন করবে?
- এটি ডেটার নির্ভুলতা এবং নির্ভুলতা নিয়ে কাজ করে।


4. মডেল প্রশিক্ষণ: 

- প্রথম তিনটি ধাপ সম্পূর্ণভাবে সম্পন্ন করার পর আমরা মডেল প্রশিক্ষণ পর্বে প্রবেশ করি।
- এটি আনুষ্ঠানিকভাবে প্রথম পদক্ষেপ যখন ডেভেলপার ডেটার ভিত্তিতে মডেলটিকে প্রশিক্ষণ দেয়।
- মডেলটি প্রশিক্ষণের জন্য, ডেটা তিনটি ভাগে বিভক্ত করা হয়েছে- প্রশিক্ষণ ডেটা, বৈধতা ডেটা এবং পরীক্ষার ডেটা।
- প্রায় 70%-80% ডেটা ট্রেনিং ডেটা সেটে যায় যা মডেল প্রশিক্ষণে ব্যবহৃত হয়।
- বৈধকরণ ডেটা ডেভেলপমেন্ট সেট বা ডেভ সেট নামেও পরিচিত এবং এটি অতিরিক্ত ফিটিং বা আন্ডারফিটিং পরিস্থিতি এড়াতে যেমন হাইপারপ্যারামিটার টিউনিং সক্ষম করার জন্য ব্যবহৃত হয়।
- হাইপারপ্যারামিটার টিউনিং হল ওভারফিটিং এবং আন্ডারফিটিং এর বিরুদ্ধে লড়াই করার জন্য ব্যবহৃত একটি কৌশল।
- মডেল মূল্যায়নের সময় বৈধতা ডেটা ব্যবহার করা হয়।
- প্রায় 10%-15% ডেটা বৈধকরণ ডেটা হিসাবে ব্যবহৃত হয়।
- বাকি 10%-15% ডেটা পরীক্ষার ডেটা সেটে যায়। পরীক্ষার ডেটা সেট মডেল প্রস্তুতির পরে পরীক্ষার জন্য ব্যবহার করা হয়।
- একটি সঠিক মডেল পেতে ডেটা বিভক্ত করার সময় ডেটা সেটগুলিকে এলোমেলো করা অত্যন্ত গুরুত্বপূর্ণ৷
- স্কিট লার্ন ইন পাইথন ব্যবহার করে ডেটা এলোমেলো করা যেতে পারে।


5. মডেল মূল্যায়ন: 

- মডেল প্রশিক্ষণের পরে, যাচাইকরণ, বা বিকাশ ডেটা মডেল মূল্যায়ন করতে ব্যবহৃত হয়।
- ডেটা পরীক্ষার জন্য সবচেয়ে সঠিক ভবিষ্যদ্বাণী পেতে আরও মডেল মূল্যায়নের জন্য ব্যবহার করা যেতে পারে।
- সংখ্যাগতভাবে নির্ভুলতা এবং নির্ভুলতা গণনা করার জন্য মডেল মূল্যায়নের পরে একটি বিভ্রান্তি ম্যাট্রিক্স তৈরি করা হয়।
- মডেল মূল্যায়নের পর, আমাদের মডেল চূড়ান্ত পর্যায়ে প্রবেশ করে যা ভবিষ্যদ্বাণী।


6. ভবিষ্যদ্বাণী:  

- ভবিষ্যদ্বাণী পর্যায়ে বিকাশকারী মডেলটি স্থাপন করে।
- মডেল স্থাপনের পরে, এটি ভবিষ্যদ্বাণী করার জন্য প্রস্তুত হয়ে যায়।
- বিল্ড মডেল সম্পর্কে আরও ভাল বোঝার জন্য প্রশিক্ষণ ডেটা এবং পরীক্ষার ডেটার উপর ভবিষ্যদ্বাণী করা হয়।
