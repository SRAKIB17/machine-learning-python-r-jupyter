3. আধা-তত্ত্বাবধানে শিক্ষা(semi-supervised learning):


##  প্রয়োজনীয় আউটপুটের উপর ভিত্তি করে শ্রেণীকরণ
মেশিন-লার্নিং কাজগুলির আরেকটি শ্রেণীকরণ দেখা দেয় যখন কেউ একটি মেশিন-লার্নিং সিস্টেমের পছন্দসই আউটপুট বিবেচনা করে:  

###### শ্রেণীবিভাগ:
যখন ইনপুটগুলিকে দুই বা ততোধিক শ্রেণীতে ভাগ করা হয়, তখন অবশ্যই একটি মডেল তৈরি করতে হবে যা এই শ্রেণীর এক বা একাধিক (মাল্টি-লেবেল শ্রেণীবিভাগ) অদেখা ইনপুট বরাদ্দ করে। এটি সাধারণত একটি তত্ত্বাবধানে মোকাবেলা করা হয়। স্প্যাম ফিল্টারিং হল শ্রেণীবিভাগের একটি উদাহরণ, যেখানে ইনপুটগুলি হল ইমেল (বা অন্যান্য) বার্তা এবং ক্লাসগুলি হল "স্প্যাম" এবং "স্প্যাম নয়"৷

###### রিগ্রেশন:
এটি একটি তত্ত্বাবধানে থাকা সমস্যা, একটি ক্ষেত্রে যখন আউটপুটগুলি বিচ্ছিন্ন না হয়ে অবিচ্ছিন্ন থাকে।

###### ক্লাস্টারিং:
যখন ইনপুটের একটি সেটকে গ্রুপে ভাগ করতে হয়। শ্রেণীবিভাগের বিপরীতে, গোষ্ঠীগুলি আগে থেকে জানা যায় না, এটি সাধারণত একটি তত্ত্বাবধানহীন কাজ করে তোলে।
