# মেশিন লার্নিং এর পরিভাষা

1. Model: মডেল একটি মডেল হল একটি নির্দিষ্ট উপস্থাপনা যা কিছু মেশিন লার্নিং অ্যালগরিদম প্রয়োগ করে ডেটা থেকে শেখা। একটি মডেলকে <b>হাইপোথিসিসও</b> বলা হয় ।
   
2. (Feature): একটি বৈশিষ্ট্য আমাদের ডেটার একটি পৃথক পরিমাপযোগ্য সম্পত্তি। সাংখ্যিক বৈশিষ্ট্যগুলির একটি সেট একটি Feature ভেক্টর দ্বারা সুবিধাজনকভাবে বর্ণনা করা যেতে পারে । Feature ভেক্টর মডেল ইনপুট হিসাবে খাওয়ানো হয়. উদাহরণস্বরূপ, একটি ফলের ভবিষ্যদ্বাণী করার জন্য, রঙ, গন্ধ, স্বাদ ইত্যাদির মতো বৈশিষ্ট্য থাকতে পারে৷ দ্রষ্টব্য: তথ্যপূর্ণ, বৈষম্যমূলক এবং স্বাধীন বৈশিষ্ট্যগুলি বেছে নেওয়া কার্যকর অ্যালগরিদমের জন্য একটি গুরুত্বপূর্ণ পদক্ষেপ৷ আমরা সাধারণত কাঁচা ডেটা থেকে প্রাসঙ্গিক বৈশিষ্ট্যগুলি বের করার জন্য একটি বৈশিষ্ট্য নিষ্কাশনকারী নিয়োগ করি।
   
3. Target (Label): একটি টার্গেট ভেরিয়েবল বা লেবেল হল সেই মান যা আমাদের মডেল দ্বারা অনুমান করা হবে। বৈশিষ্ট্য বিভাগে আলোচনা করা ফলের উদাহরণের জন্য, প্রতিটি সেটের ইনপুট সহ লেবেল হবে আপেল, কমলা, কলা ইত্যাদি ফলের নাম।
   
4. Training:  প্রশিক্ষণের ধারণাটি হল ইনপুট (বৈশিষ্ট্য) এবং এর প্রত্যাশিত আউটপুট (লেবেল) এর একটি সেট দেওয়া, তাই প্রশিক্ষণের পরে, আমাদের কাছে একটি মডেল (অনুমান) থাকবে যা তারপর প্রশিক্ষণপ্রাপ্ত বিভাগগুলির মধ্যে একটিতে নতুন ডেটা ম্যাপ করবে।
   
5. Prediction: ভবিষ্যদ্বাণী আমাদের মডেল প্রস্তুত হয়ে গেলে, এটিকে ইনপুটের একটি সেট খাওয়ানো যেতে পারে যা এটি একটি পূর্বাভাসিত আউটপুট (লেবেল) প্রদান করবে। তবে নিশ্চিত করুন যে মেশিনটি অদেখা ডেটাতে ভাল পারফর্ম করে, তবেই আমরা বলতে পারি মেশিনটি ভাল পারফর্ম করে।
