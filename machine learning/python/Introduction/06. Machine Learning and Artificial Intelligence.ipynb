**কৃত্রিম বুদ্ধিমত্তা** (AI) এবং **মেশিন লার্নিং (ML)** কম্পিউটার বিজ্ঞানের বিস্তৃত ক্ষেত্রের মধ্যে দুটি ঘনিষ্ঠভাবে সম্পর্কিত কিন্তু স্বতন্ত্র ক্ষেত্র। AI হল একটি শৃঙ্খলা যা বুদ্ধিমান মেশিন তৈরির উপর ফোকাস করে যে কাজগুলি করতে পারে যেগুলি সাধারণত মানুষের বুদ্ধিমত্তার প্রয়োজন হয়, যেমন ভিজ্যুয়াল উপলব্ধি, বক্তৃতা শনাক্তকরণ, সিদ্ধান্ত গ্রহণ এবং প্রাকৃতিক ভাষা প্রক্রিয়াকরণ। এতে অ্যালগরিদম এবং সিস্টেমের বিকাশ জড়িত যা ইনপুট ডেটার উপর ভিত্তি করে যুক্তি, শিখতে এবং সিদ্ধান্ত নিতে পারে।

অন্যদিকে**মেশিন লার্নিং**িং (এমএল) হল AI এর একটি সাবফিল্ড যা স্পষ্টভাবে প্রোগ্রাম করা ছাড়াই ডেটা থেকে শেখার জন্য মেশিনগুলিকে শেখানোর সাথে জড়িত। ML অ্যালগরিদমগুলি ডেটার নিদর্শন এবং প্রবণতাগুলি সনাক্ত করতে পারে এবং ভবিষ্যদ্বাণী এবং সিদ্ধান্ত নিতে তাদের ব্যবহার করতে পারে। ML ভবিষ্যদ্বাণীমূলক মডেল তৈরি করতে, ডেটা শ্রেণীবদ্ধ করতে এবং নিদর্শনগুলি সনাক্ত করতে ব্যবহৃত হয় এবং এটি অনেক AI অ্যাপ্লিকেশনের জন্য একটি অপরিহার্য হাতিয়ার।
