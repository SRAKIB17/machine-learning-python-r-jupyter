# মেশিন লার্নিং সমস্যার ধরন

### একটি শেখার সিস্টেমে উপলব্ধ "সংকেত" বা "প্রতিক্রিয়া" শেখার প্রকৃতির ভিত্তিতে


#### তত্ত্বাবধানে শিক্ষা (Supervised learning):

মডেল বা অ্যালগরিদম উদাহরণ ইনপুট এবং তাদের পছন্দসই আউটপুট সহ উপস্থাপন করা হয় এবং তারপর ইনপুট এবং আউটপুটের মধ্যে প্যাটার্ন এবং সংযোগ খুঁজে পায়। লক্ষ্য হল একটি সাধারণ নিয়ম শেখা যা ইনপুটগুলিকে আউটপুটগুলিতে ম্যাপ করে। যতক্ষণ না মডেলটি প্রশিক্ষণের ডেটার উপর কাঙ্খিত স্তরের নির্ভুলতা অর্জন করে ততক্ষণ পর্যন্ত প্রশিক্ষণ প্রক্রিয়া চলতে থাকে। 

<b>বাস্তব জীবনের কিছু উদাহরণ হল:</b><br/>

1. ছবির শ্রেণীবিভাগ: আপনি ছবি/লেবেল দিয়ে প্রশিক্ষণ দেন। তারপরে ভবিষ্যতে, আপনি একটি নতুন চিত্র দেবেন এই প্রত্যাশায় যে কম্পিউটারটি নতুন বস্তুটিকে চিনবে।
2. বাজারের পূর্বাভাস/অবস্থান: আপনি ঐতিহাসিক বাজারের তথ্য দিয়ে কম্পিউটারকে প্রশিক্ষণ দেন এবং কম্পিউটারকে ভবিষ্যতে নতুন মূল্যের পূর্বাভাস দিতে বলুন।

### তত্ত্বাবধানহীন শিক্ষা (Unsupervised learning):

শেখার অ্যালগরিদমকে কোনো লেবেল দেওয়া হয় না, এটির ইনপুটে গঠন খুঁজে পেতে এটিকে নিজের উপর রেখে দেয়। এটি বিভিন্ন গ্রুপে জনসংখ্যার ক্লাস্টারিংয়ের জন্য ব্যবহৃত হয়। তত্ত্বাবধানহীন শিক্ষা নিজেই একটি লক্ষ্য হতে পারে (ডেটাতে লুকানো নিদর্শনগুলি আবিষ্কার করা)।

1. ক্লাস্টারিং: আপনি কম্পিউটারকে অনুরূপ ডেটা ক্লাস্টারে আলাদা করতে বলুন, এটি গবেষণা এবং বিজ্ঞানে অপরিহার্য।
2. হাই-ডাইমেনশন ভিজ্যুয়ালাইজেশন: হাই-ডাইমেনশন ডেটা ভিজ্যুয়ালাইজ করতে আমাদের সাহায্য করতে কম্পিউটার ব্যবহার করুন।
3. জেনারেটিভ মডেল: একটি মডেল আপনার ইনপুট ডেটার সম্ভাব্যতা বিতরণ ক্যাপচার করার পরে, এটি আরও ডেটা তৈরি করতে সক্ষম হবে। এটি আপনার ক্লাসিফায়ারকে আরও শক্তিশালী করতে খুব কার্যকর হতে পারে।

### আধা-তত্ত্বাবধানে শেখার সমস্যা(Semi-supervised learning):
যেখানে আপনার কাছে প্রচুর পরিমাণে ইনপুট ডেটা থাকে এবং শুধুমাত্র কিছু ডেটা লেবেলযুক্ত থাকে, সেগুলিকে আধা-তত্ত্বাবধানে শেখার সমস্যা বলে। এই সমস্যাগুলি তত্ত্বাবধান করা এবং তত্ত্বাবধানহীন উভয় শিক্ষার মধ্যে বসে। উদাহরণস্বরূপ, একটি ফটো সংরক্ষণাগার যেখানে শুধুমাত্র কিছু ছবি লেবেলযুক্ত, (যেমন কুকুর, বিড়াল, ব্যক্তি) এবং বেশিরভাগই লেবেলবিহীন।

### শক্তিবৃদ্ধি শিক্ষা(Reinforcement learning): 
একটি কম্পিউটার প্রোগ্রাম একটি গতিশীল পরিবেশের সাথে যোগাযোগ করে যেখানে এটি একটি নির্দিষ্ট লক্ষ্য (যেমন একটি যানবাহন চালানো বা প্রতিপক্ষের বিরুদ্ধে একটি গেম খেলা) সম্পাদন করতে হবে। প্রোগ্রামটিকে পুরষ্কার এবং শাস্তির পরিপ্রেক্ষিতে প্রতিক্রিয়া প্রদান করা হয় কারণ এটি তার সমস্যার স্থান নেভিগেট করে।

 ## তত্ত্বাবধানে শিক্ষার দুটি সবচেয়ে সাধারণ ব্যবহার হল: 

#### শ্রেণীবিভাগ(Classification):
ইনপুটগুলিকে দুই বা ততোধিক শ্রেণীতে বিভক্ত করা হয়, এবং অবশ্যই একটি মডেল তৈরি করতে হবে যা এই শ্রেণীর এক বা একাধিক (মাল্টি-লেবেল শ্রেণীবিভাগ) অদেখা ইনপুট বরাদ্দ করে এবং কিছু একটি নির্দিষ্ট শ্রেণীর অন্তর্গত কিনা তা ভবিষ্যদ্বাণী করে। এটি সাধারণত একটি তত্ত্বাবধানে মোকাবেলা করা হয়। শ্রেণীবিভাগ মডেল দুটি গ্রুপে শ্রেণীবদ্ধ করা যেতে পারে: বাইনারি শ্রেণীবিভাগ এবং মাল্টিক্লাস শ্রেণীবিভাগ। স্প্যাম ফিল্টারিং হল বাইনারি শ্রেণীবিভাগের একটি উদাহরণ, যেখানে ইনপুটগুলি হল ইমেল (বা অন্য) বার্তা এবং ক্লাসগুলি হল "স্প্যাম" এবং "স্প্যাম নয়"৷

### রিগ্রেশন(Regression):
এটি একটি তত্ত্বাবধানে শিক্ষার সমস্যা, যা একটি সাংখ্যিক মান ভবিষ্যদ্বাণী করে এবং আউটপুটগুলি বিচ্ছিন্ন না হয়ে অবিচ্ছিন্ন থাকে। উদাহরণস্বরূপ, ঐতিহাসিক তথ্য ব্যবহার করে স্টক মূল্য ভবিষ্যদ্বাণী করা।

## সবচেয়ে সাধারণ তত্ত্বাবধানহীন শিক্ষা (Unsupervised learning) হল

### ক্লাস্টারিং(Clustering): 
এখানে, ইনপুটগুলির একটি সেটকে গ্রুপে ভাগ করতে হবে। শ্রেণীবিভাগের বিপরীতে, গোষ্ঠীগুলি আগে থেকে জানা যায় না, এটি সাধারণত একটি তত্ত্বাবধানহীন কাজ করে তোলে। আপনি নীচের উদাহরণে দেখতে পাচ্ছেন, প্রদত্ত ডেটাসেট পয়েন্টগুলি লাল, সবুজ এবং নীল রঙের দ্বারা শনাক্তযোগ্য গ্রুপে বিভক্ত করা হয়েছে।

### ঘনত্ব অনুমান(Density estimation): 
কাজটি হল কিছু জায়গায় ইনপুটগুলির বিতরণ খুঁজে বের করা।

### মাত্রিকতা হ্রাস(Dimensionality reduction):
এটি একটি নিম্ন-মাত্রিক স্থানে ম্যাপ করে ইনপুটগুলিকে সরল করে। টপিক মডেলিং একটি সম্পর্কিত সমস্যা, যেখানে একটি প্রোগ্রামকে মানব ভাষার নথিগুলির একটি তালিকা দেওয়া হয় এবং কোন নথিগুলি অনুরূপ বিষয়গুলি কভার করে তা খুঁজে বের করার দায়িত্ব দেওয়া হয়।
