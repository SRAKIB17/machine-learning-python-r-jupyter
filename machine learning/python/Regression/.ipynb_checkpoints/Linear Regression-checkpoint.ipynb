## রিগ্রেশন
যখন ভেরিয়েবলের মধ্যে সম্পর্ক খুঁজে বের করার চেষ্টা করেন তখন রিগ্রেশন শব্দটি ব্যবহার করা হয়।
মেশিন লার্নিং এবং পরিসংখ্যানগত মডেলিং-এ, সেই সম্পর্ক ভবিষ্যতের ঘটনাগুলির ফলাফলের পূর্বাভাস দিতে ব্যবহৃত হয়।
লিনিয়ার রিগ্রেশন
রৈখিক রিগ্রেশন ডেটা-বিন্দুর মধ্যে সম্পর্ক ব্যবহার করে সেগুলির মধ্যে একটি সরল রেখা আঁকতে পারে। এই লাইনটি ভবিষ্যতের মান অনুমান করতে ব্যবহার করা যেতে পারে। মেশিন লার্নিং-এ, ভবিষ্যতের ভবিষ্যদ্বাণী করা খুবই গুরুত্বপূর্ণ। পাইথনে ডেটা-পয়েন্টের মধ্যে সম্পর্ক খুঁজে বের করার এবং রৈখিক রিগ্রেশনের একটি রেখা আঁকার পদ্ধতি রয়েছে।

x-অক্ষের মান এবং y-অক্ষের মানগুলির মধ্যে সম্পর্ক কেমন তা জানা গুরুত্বপূর্ণ, যদি কোনও সম্পর্ক না থাকে তবে রৈখিক রিগ্রেশন কোনও কিছুর পূর্বাভাস দিতে ব্যবহার করা যায় না। এই সম্পর্ক - পারস্পরিক সম্পর্কের সহগ  r বলা হয় । This relationship - the coefficient of correlation - is called r.
r = মান -1 থেকে 1 পর্যন্ত, যেখানে 0 মানে কোনো সম্পর্ক নেই, এবং 1 (এবং -1) মানে 100% সম্পর্কিত.
দ্রষ্টব্য: ফলাফল -0.76 দেখায় যে একটি সম্পর্ক আছে, নিখুঁত নয়, তবে এটি নির্দেশ করে যে আমরা ভবিষ্যতের ভবিষ্যদ্বাণীতে লিনিয়ার রিগ্রেশন ব্যবহার করতে পারি।
